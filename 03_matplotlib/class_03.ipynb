```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.array([1,2,3,4,5])
y=x**2
```

```python
plt.plot(x, y,marker="o",color="r")
plt.plot(x,x**3,marker="*",color="y")
plt.title("this is a simple line plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend(["line1","line2"])
plt.show()
```

```python
emni=np.arange(10)
x=np.random.normal(170,10,250)
players="a b c d".split()
run=[51,87,45,67]
plt.figure(figsize=(20,30))
plt.subplot(3,2,1)
plt.hist(x)
plt.subplot(3,2,2)
plt.scatter(emni,emni**2)
plt.subplot(3,2,3)
plt.bar(players,run,color='b')
plt.subplot(3,2,4)
plt.barh(players,run,color='r')
plt.subplot(3,2,5)
plt.pie(run,explode=(0,0,0,0.1),labels=players,shadow=True,startangle=90,autopct='%.1f')
plt.subplot(3,2,6)
plt.plot(emni,emni**2)
plt.show()
```
